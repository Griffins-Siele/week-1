### Activity 2: Simple Shopping List

Create a program that:

### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# Record temperatures for 5 days
with open('temps.txt', 'w') as file:
    for day in range(1, 6):# Day 1 to Day 5
        temp = float(input(f"Enter temperature for day {day}: "))
        file.write(f"Day {day}: {temp}°C\n")

# Read temperatures and calculate average
with open('temps.txt', 'r') as file:
    lines = file.readlines()

temps = []
for line in lines:
# Extract temperature value
    temp_value = float(line.split(":")[1].replace("°C", "").strip())
    temps.append(temp_value)

# Calculate average temperature
average_temp = sum(temps) / len(temps)
print(f"Average temperature over 5 days: {average_temp:.2f}°C")
`````

### **Activity 4: Movie Collection Tuple** 


Create a tuple with 3 movies and their release years:
movies = (("Inception", 2010), ("Matrix", 1999), ("Interstellar", 2014))

Print:
- The oldest movie
- All movie titles
- Try to change a year (show it causes an error)

---

`````python
# create a tuple for movies and years
movies = (("Inception", 2010), ("The Matrix", 1999), ("Interstellar", 2014), ("The Godfather", 1972))

print('All Movies:')
for movie in movies:
    print(f"{movie[0]}")

# find the oldest movie
oldest = min(movies, key=lambda m: m[1])
print("\nOldest movie:", oldest[0], "released in", oldest[1])

# try changing a year (will cause an error)
try:
    movies[0][1] = 2025  # Attempt to change year
except TypeError as e:
    print("\nError:", e)
`````

### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

Find and print:
- Hobbies both enjoy (intersection)
- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
# create the sets
alex_hobbies = {'reading', 'gaming', 'cooking', 'hiking'}
sam_hobbies = {'gaming', 'hiking', 'photohraphy', 'music'}

# common hobbies
common_hobbies = alex_hobbies & sam_hobbies
print("Common hobbies:", common_hobbies)

# hobbies only alex enjoys
only_alex = alex_hobbies - sam_hobbies
print("Hobbies only Alex enjoys:", only_alex)

# all unique hobbies
all_hobbies = alex_hobbies | sam_hobbies
print("All unique hobbies:", all_hobbies)
`````


### **Activity 6: Simple Grade Book Dictionary** 

Create a dictionary for 3 students and their scores:
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

Perform these tasks:
- Add a new student "Diana" with score 88
- Update Bob's score to 95
- Print all students with scores above 80
- Calculate and print the average score

---

`````python
# create a dictionary
grades = {'Alice': 85, 'Bob': 92, 'Charlie': 78}
# add new student
grades['Diana'] = 88 
# update Bob's grade 
grades['Bob'] = 95 

# print students with scores above 80
print("Students with scores above 80:")
for student, score in grades.items():
    if score > 80:
        print(f'{student}: {score}')

# calculate the average score
average = sum(grades.values()) / len(grades)
print(f'Average score: {average:.2f}')
`````

### Activity 9: Finding "Dream" Sentences in Les_Brown.txt

From the `Les_Brown.txt` file, read the content and identify all sentences that contain the word "dream". Write these sentences to a new file called `dream_sentences.txt`.

**Steps:**
1. Open and read the content of `Les_Brown.txt`.
2. Split the text into sentences.
3. Filter out sentences that contain the word "dream" (case-insensitive).
4. Write these sentences into the new file `dream_sentences.txt`.
<hr>

`````python
with open('Les_Brown.txt', 'r') as file:
    content = file.read()
# split content into sentences
sentences = content.split('.')
print("Total sentences in file:", len(sentences))
#filter sentences containing 'dream'
dream_sentences = [s.strip() for s in sentences if 'dream' in s.lower()]
print("Sentences containing 'dream':")
for s in dream_sentences:
    print(s)
`````

### Activity 11: Adding Line Numbers to Python Zen

Create a Python program that reads the content from the `python_zen.txt` file and adds line numbers to each line. Each line in the output file should start with a line number, followed by a dot and a space, and then the original line from the input file. Skip numbering the first line as it is the title. Save the result into a new file called `numbered_zen.txt`.

**Steps:**
1. Open the `python_zen.txt` file for reading.
2. Skip the first line (which is the title) when numbering.
3. For each subsequent line, prepend the line number followed by a dot and a space.
4. Write the modified content to a new file called `numbered_zen.txt`.

<hr>

`````python
with open('python_zen.txt', 'r') as file:
    content = file.read()
    print(content)
# skip the first line when numbering
    lines = content.splitlines()[1:]
    for idx, line in enumerate(lines, start=1):
        print(f"{idx}: {line}")
# write the modified content to a new file
with open('numbered_zen.txt', 'w') as file:
    for idx, line in enumerate(lines, start=1):
        file.write(f"{idx}: {line}\n" )
`````
