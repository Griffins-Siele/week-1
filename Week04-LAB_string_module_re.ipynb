3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
# Function to convert password characters to ASCII numbers
def password_to_ascii(password):
    ascii_string = ""
    for char in password:
        ascii_string += str(ord(char))
    return ascii_string

# Example usage
password = input("Enter your password: ")
print("ASCII version:", password_to_ascii(password))
````

4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def validate_password(password):
    # Check if password contains space
    if " " in password:
        print("Error: Password should not contain spaces.")
        return

    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in "@#$%^&*()_+-=!?<>/" for char in password)

    # Print validation results
    print("Contains uppercase:", has_upper)
    print("Contains lowercase:", has_lower)
    print("Contains digit:", has_digit)
    print("Contains special character:", has_special)

    # If all conditions are True
    if has_upper and has_lower and has_digit and has_special:
        print(True)
    else:
        print(False)


# Example usage
password = input("Enter your password: ")
validate_password(password)
````

9. (Optional)
Basic Regex for Pattern Matching
Import the re module and use it to search for an email address in a string.
Use regex to find all the words that start with a capital letter in a sentence.

````python
import re

text = "Contact us at info@example.com or reach out to Griffins and Siele for help."

# Search for an email address
email = re.search(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', text)
if email:
    print("Email found:", email.group())
else:
    print("No email found.")

# Find all words that start with a capital letter
capital_words = re.findall(r'\b[A-Z][a-z]*\b', text)
print("Words starting with capital letters:", capital_words)
````

10. Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
def count_characters(s):
    lower = upper = digits = special = 0
    for char in s:
        if char.islower():
            lower += 1
        elif char.isupper():
            upper += 1
        elif char.isdigit():
            digits += 1
        else:
            special += 1
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digits)
    print("Special characters:", special)

# Example usage
example_password = "P@#yn26at^&i5ve"
count_characters(example_password)
````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
def validate_password(password):
    if " " in password:
        print("Error: Password should not contain spaces.")
        return False

    if len(password) <= 8:
        print("Password must be more than 8 characters.")
        return False

    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in "@#$%^&*()_+-=!?<>/" for char in password)

    if not has_upper:
        print("Password must contain at least one uppercase letter.")
    if not has_lower:
        print("Password must contain at least one lowercase letter.")
    if not has_digit:
        print("Password must contain at least one number.")
    if not has_special:
        print("Password must contain at least one special character.")

    if has_upper and has_lower and has_digit and has_special and len(password) > 8:
        print("Valid password.")
        return True
    else:
        return False

# Ask user for password input
user_password = input("Enter your password: ")
validate_password(user_password)
````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
# Griffins Siele (M01087820)
import random

special_chars = "@#$%^&*!"

def generate_secret_word(first_name, surname):
    secret = first_name[:3] + surname[-2:]  # first 3 letters of first name + last 2 of surname
    secret += random.choice(special_chars)  # append a random special character
    return secret

# Ask user for input
first_name = input("Enter your first name: ").strip()
surname = input("Enter your surname: ").strip()

# Generate and print secret word
print("Your secret word is:", generate_secret_word(first_name, surname))
````

14. Advanced Student Number Validation: Validate the student number to ensure:
- It starts with "M00".
- It contains exactly 9 characters.
- It has no special characters or spaces

````python
def validate_student_number(student_number):
    if not student_number.startswith("M00"):
        return False
    if len(student_number) != 9:
        return False
    if not student_number.isalnum():
        return False  # checks for letters and digits only (no special characters or spaces)
    return True

# Example usage
student_number = input("Enter your student number: ").strip()
if validate_student_number(student_number):
    print("Valid student number.")
else:
    print("Invalid student number.")
````
